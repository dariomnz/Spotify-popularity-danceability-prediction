```python
import pandas as pd 

df = pd.read_csv("../dataset/tracks.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Crear la matriz de correlación
corr = df.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr))
fig, ax = plt.subplots(figsize=(12, 12))
# Visualizar la matriz de correlación con seaborn
sns.set(style='darkgrid')
sns.heatmap(corr, annot=True, fmt='.2f', cmap="coolwarm",linewidths=0.5, square=True, mask=mask)
plt.savefig("../documentation/images/matriz_correlacion.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15, 6))
plt.subplot(1,2,1)
sns.histplot(df['popularity']).set_title('Popularity Distribution')
plt.subplot(1,2,2)
sns.histplot(df['danceability']).set_title('Danceability Distribution')
plt.savefig("../documentation/images/histogramas.png")
```

```python
fig, ax = plt.subplots(figsize = (15, 6))
sns.scatterplot(x='loudness', y='popularity', data=df, color='blue', alpha=0.3)
plt.savefig("../documentation/images/popularity_loudness_scatterplot.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15, 6))
sns.scatterplot(x='danceability', y='speechiness', data=df, color='blue', alpha=0.3)
plt.savefig("../documentation/images/danceability_speechiness_scatterplot.png")
plt.show()
```
